```python
# Creating a function that applys a kernel
def applykernel(image, kernel):
    
    output = cv2.filter2D(image,-1,kernel)
    
    return output
```

```python
# kernel 1
kernel_1 = np.ones((3, 3)) * (1.5 /9)
```

```python
display_1 = applykernel(image1, kernel_1)
plt.imshow(display_1)    
```

```python
# HSV Color-Space Histogram
color = ('r','g','b')
for i,col in enumerate(color):
    histr = cv2.calcHist([display_1],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,768])
    plt.ylim([0,100000])
plt.show()
```

```python
# kernel 2
gaussianKernel = np.array((1, 2, 1, 2, 4, 2, 1, 2, 1))
gaussianKernel = np.reshape(gaussianKernel, (3, 3)) * (1/16)
```

```python
display_2 = applykernel(image1, gaussianKernel)
plt.imshow(display_2)
```

```python
# display_1 appears brighter than display_2
# hence kernel_1 does a better job.
```

```python
# increasing the size of the kernel_1
kernel_3 = np.ones((5, 5)) * (2.5 /25)
display_3 = applykernel(image1, kernel_3)
plt.imshow(display_3)
```

```python
# checking the the image hist of display_3
color = ('r','g','b')
for i,col in enumerate(color):
    histr = cv2.calcHist([display_3],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,768])
    plt.ylim([0,100000])
plt.show()
```

```python
# increasing the size of the kernel_3
kernel_4 = np.ones((11,11)) * (5.5 /121)
display_4 = applykernel(image1, kernel_4)
plt.imshow(display_4)
```

```python
# checking the the image hist of display_4
color = ('r','g','b')
for i,col in enumerate(color):
    histr = cv2.calcHist([display_4],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,768])
    plt.ylim([0,100000])
plt.show()
```

```python
alpha = 1.5 # Simple contrast control
beta = 1.5    # Simple brightness control
```

```python
new_image = np.zeros(display_4.shape, display_3.dtype)
```

```python
for y in range(display_4.shape[0]):
    for x in range(display_4.shape[1]):
        for c in range(display_4.shape[2]):
            new_image[y,x,c] = np.clip(alpha*display_4[y,x,c] + beta, 0, 255)
```

```python
plt.imshow(new_image)
```

```python
# checking the the image hist of new_image
color = ('r','g','b')
for i,col in enumerate(color):
    histr = cv2.calcHist([new_image],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,768])
    plt.ylim([0,100000])
plt.show()
```

```python
gamma = 1.0
lookUpTable = np.empty((1,256), np.uint8)
for i in range(256):
    lookUpTable[0,i] = np.clip(pow(i / 255.0, gamma) * 255.0, 0, 255)
res2 = cv2.LUT(display_4, lookUpTable)
```

```python
plt.imshow(res2)
```

```python
# checking the the image hist of res2
color = ('r','g','b')
for i,col in enumerate(color):
    histr = cv2.calcHist([res2],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,768])
    plt.ylim([0,100000])
plt.show()
```
